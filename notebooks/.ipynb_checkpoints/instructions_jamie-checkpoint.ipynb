One of the key challenges in any data set are missing values. Lets start by checking which columns contain missing values.

````python
# Confirm Loan_ID unique.  If so, it can be dropped for EDA

if df.shape[0]==df['Loan_ID'].nunique():
    print(df.shape)
    df.drop(columns=['Loan_ID'], inplace=True)
    print(df.shape)
````

````python
# Check for nulls:
df_nulls = df.isnull().sum()
perc = round((df_nulls/df.isnull().count())*100,1)
df_missing_data = pd.concat(
    [df_nulls, perc]
  , axis=1
  , keys=['null_cnt', 'null_%']
  , verify_integrity= True
)
# Create Meta data table of Variable details:
meta = df_missing_data.copy()
meta.reset_index(inplace=True, names= 'variable')
meta['dtype'] = list(df.dtypes)

print(f'Shape of data: {df.shape}')
print(f'Nulls in data:')
meta.sort_values('null_%', ascending=False)
````

````python
# Modularize the above code into a function:
def null_proportions(df, printout=False):
    null = df.isnull().sum()
    perc = round((null/df.isnull().count())*100,1)
    null_df = pd.concat([null, perc]
                        , axis=1
                        , keys=['null_cnt', 'null_%']
                        , verify_integrity= True)
    null_df.reset_index(inplace= True, names= 'variable')
    # Add data type column:
    null_df['dtype'] = list(df.dtypes)
    
    if printout== True:
        print(f'Shape of data: {df.shape}')
        print(f'Nulls remaining:')
        display(null_df[null_df['null_%']>0
            ].sort_values('null_%', ascending=False))
    return null_df
````

````python
df_null = null_proportions(df, True)
````

Modularize the code above as a function:

````python
def descriptive_statistics(df):
    num_cols = list(df.dtypes[df.dtypes != 'object'].index)
    dict = {}
    dict['variable'] = list(num_cols)
    dict['sum']      = list(df[num_cols].sum(skipna=True))
    dict['count']    = list(df[num_cols].count())
    dict['mean']     = list(round(df[num_cols].mean(skipna=True), 1))
    dict['std']      = list(round(df[num_cols].std(skipna=True), 1))
    dict['min']      = list(df[num_cols].min(skipna=True))
    dict['median']   = list(df[num_cols].median(skipna=True))
    dict['max']      = list(df[num_cols].max(skipna=True))
    return pd.DataFrame(dict)
````


3. Tip: Can you see a possible skewness in the data by comparing the mean to the median, i.e. the 50% figure of a feature.



````python
# Will use the Shapiro Wilk Normality test to test the 
# H0: The distribution is normal: mean = median
from scipy import stats

def shapiro_normality(df, col_name):
    from scipy import stats
    stat, p = stats.shapiro(df[col_name])
    h0_keep = (f"{col_name} Normality:\n  Evidence insufficient to reject H0 as p < 0.05 = False:  p:{'%0.5f' % p}, stat: {round(stat, 3)}")
    h0_rej = (f"{col_name} Normality:\n  Likely Gaussian, H0 rejected: p:{'%0.5f' % p}, stat: {round(stat, 3)}")
    return h0_rej if p < 0.05 else h0_keep
````

````python
print(shapiro_normality(df, 'ApplicantIncome'))
print(shapiro_normality(df, 'CoapplicantIncome'))
````

````python
def hist_compare(df, col1, col2):
  fig, axes = plt.subplots(1, 2, figsize = (10,3))
  # Plot 0:
  axes[0].hist(df[col1], bins=40, alpha=0.4, label="bins=40", color= 'green')
  axes[0].hist(df[col1], bins=80, alpha=0.4, label="bins=80", color= 'blue')
  axes[0].legend(loc='upper right')
  axes[0].set_title(f'Histogram of {col1}')
  axes[0].set_xlabel(col1)
  axes[0].set_ylabel('Count Frequency')

  # Plot 1:
  axes[1].hist(df[col2], bins=40, alpha=0.4, label="bins=40", color= 'green')
  axes[1].hist(df[col2], bins=80, alpha=0.4, label="bins=80", color= 'blue')
  axes[1].legend(loc='upper right')
  axes[1].set_title(f'Histogram of {col2}')
  axes[1].set_xlabel(col2)
  axes[1].set_ylabel('Count Frequency')
  plt.savefig(f'../images/hist_compare_{col1}_{col2}.png')
  plt.show()
````

````python
print(hist_compare(df, 'LoanAmount', 'log_LoanAmount'))
print(hist_compare(df, 'ApplicantIncome', 'log_ApplicantIncome'))
print(hist_compare(df, 'CoapplicantIncome', 'log_CoapplicantIncome'))
print(hist_compare(df, 'combined_incomes', 'log_combined_incomes'))
````

Define a function `mean_loan_by_proportioned_cat_vbls(df)` to plot line charts of multi-indexed pivot tables to illustrate the mean Loan amounts as a function of proportionately represented categorical variable, for loans approved (`loan_status` = '`Y`') in green line, and not approved (`loan_status` = '`N`') in red.

````python
def linechart_avg_loan_by_categ_ratio(df):
    df_in = df.drop(columns=['Loan_Status'])
    cat_cols = list(df_in.dtypes[df_in.dtypes == 'object'].index)
    loan_status = df['Loan_Status']
    for col in cat_cols:
        # Calculate col proportions as percentage
        col_perc = pd.DataFrame(round(df[col].value_counts(normalize=True)*100, 0)).reset_index()
        col_perc.columns = [col, f'%_{col}']
        df = pd.merge(left=df, right=col_perc, how='inner', on=[col])

        piv = round(pd.pivot_table(
                df
            , values = 'LoanAmount'
            , index = [col, f'%_{col}']
            , columns =  loan_status
            , aggfunc = 'mean'
        ), 2) # Round to 2 dec points

        title=f'Mean Loan $ by {col} as % Loan Status'
        piv.plot.line(
            title= title
            , color= {'N': 'red', 'Y': 'green'}
            , figsize= (14, 3)
        )
        plt.savefig(f'../images/linechart_avg_loan_by_{col}_ratio.png')
    
        print(f'{title}:\n {piv}\n')
````

````python
linechart_avg_loan_by_categ_ratio(df)
````

````python
def boxplot_avg_loan_by_categ_ratio(df):
    df_in = df.drop(columns=['Loan_Status'])
    cat_cols = list(df_in.dtypes[df_in.dtypes == 'object'].index)
    loan_status = df['Loan_Status']
    for col in cat_cols:

        # Calculate col proportions as percentage
        col_perc = pd.DataFrame(round(df[col].value_counts(normalize=True)*100, 0)).reset_index()
        col_perc.columns = [col, f'%_{col}']
        df_per = pd.merge(left=df, right=col_perc, how='inner', on=[col])
        x_var = f'%_{col}'

        # Self_Employed
        data = df_per[['LoanAmount', x_var]]
        f, ax = plt.subplots(figsize=(4, 3))
        fig = sns.boxplot(x=x_var, y="LoanAmount", data=data)
        fig.axis(ymin=-0, ymax=700)

        plt.savefig(f'../images/boxplot_avg_loan_by_{col}_ratio.png')


boxplot_avg_loan_by_categ_ratio(df)
````

## PART 5: Building a Predictive Model:

````python
from numpy import ravel                                   # For matrices
from sklearn.model_selection import train_test_split      # train/test splits
from sklearn.ensemble import RandomForestClassifier       # rf classifier
from sklearn.neighbors import KNeighborsClassifier        # knn classifier
from sklearn.decomposition import PCA                     # Feature selector
from sklearn.feature_selection import VarianceThreshold   # Feature selector
from sklearn.feature_selection import SelectKBest         # Feature selector
from sklearn.pipeline import Pipeline, FeatureUnion       # For setting up pipeline
from sklearn.preprocessing import (Normalizer    # Various pre-processing steps
                , StandardScaler
                , MinMaxScaler
                , PowerTransformer
                , MaxAbsScaler
                , LabelEncoder)
from sklearn.model_selection import GridSearchCV           # For optimization
````

## PART 6: Using Pipeline
If you didn't use pipelines before, transform your data prep, feat. engineering and modeling steps into Pipeline. It will be helpful for deployment.

The goal here is to create the pipeline that will take one row of our dataset and predict the probability of being granted a loan.

`pipeline.predict(x)`

````python
# Build pipeline: 
pipe = Pipeline([
    ('scaler', StandardScaler())
  , ('selector', VarianceThreshold())
  , ('classifier', KNeighborsClassifier())
])
 
pipe.fit(X_train, y_train)
````

````python
print(f'Training set score: {str(pipe.score(X_train,y_train))}')
print(f'Test set score: {str(pipe.score(X_test,y_test))}')
````

[TOP OF PAGE](#january-20-2023)/[TOC](#table-of-contents)

### PARAMETER GRID SEARCH:

````python
parameters = {
      'scaler': [
            StandardScaler()
          , MinMaxScaler()
          , Normalizer()
          , MaxAbsScaler()
      ]      
      , 'selector__threshold': [0, 0.001, 0.01]
      , 'classifier__n_neighbors': [1, 3, 5, 7, 10]
      , 'classifier__p': [1, 2]
      , 'classifier__leaf_size': [1, 5, 10, 15]
}
 
grid = GridSearchCV(pipe, parameters, cv=2).fit(X_train, y_train)
# grid = GridSearchCV(pipe, parameters, n_jobs = -1, verbose=10, refit=True)    
# grid.fit(X_train, y_train)
 
print(f'Training set score: {str(grid.score(X_train, y_train))}')
print(f'Test set score: {str(grid.score(X_test, y_test))}')
````

````python
# Find the best set of parameters
best_params = grid.best_params_
print(best_params)
````

````python
# Stores the optimum model in best_pipe
best_pipe = grid.best_estimator_
print(best_pipe)
````

````python
result_df = pd.DataFrame.from_dict(grid.cv_results_, orient='columns')
list(result_df.columns)
````
